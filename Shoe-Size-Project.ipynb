```python
# import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import pickle

```

```python
# next we load the data
#load data
df = pd.read_csv("Shoe-size.csv")

df.head()
```

```python
# For Gender #1 = Male;
#2 =Female
```

### Data Cleaning

```python
df.min()
```

```python
#next we drop height that is less than 100 by writing the following code

df = df[df['Height'] > 100]
```

```python
# check data shape
df.shape
```

```python
# split the data into test and train
y = df['Shoe_size'] # y is a dependent variable
X = df.drop('Shoe_size', axis=1) # X is independent variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
# check accuracy
lr.score(X_test,y_test)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
#finally we save model as a pickle file

filename = 'model.pkl'
pickle.dump(lr, open(filename, 'wb'))
```
